# C-->Q
# t test for 2 independent variables
# f test for more than 2 independent variables


```python
from scipy.stats import ttest_ind
```

```python
# t-test for two inependent sample (C->Q)
# 2-sample t-test 
gender_M = data[data['Gender']=='Male']['MonthlyIncome']
gender_F = data[data['Gender']=='Female']['MonthlyIncome']
ttest_ind(gender_M,gender_F, axis = 0, nan_policy = 'omit')
```

```python
# More thab Two independent samples (C->Q)
# One way F test - Anova

temp = data[['MonthlyIncome','BusinessTravel']]
grps = pd.unique(temp['BusinessTravel'])
d_data = {grp:data['MonthlyIncome'][data.BusinessTravel == grp] for grp in grps}
#print(d_data)
stats.f_oneway(d_data['Travel_Rarely'],d_data['Travel_Frequently'],d_data['Non-Travel'])

```

```python
temp = data[['MonthlyIncome','Department']]
grps = pd.unique(temp['Department'])
d_data = {grp:data['MonthlyIncome'][data.Department == grp] for grp in grps}
#print(d_data)
stats.f_oneway(d_data['Research & Development'],d_data['Sales'],d_data['Human Resources'])
```

```python
data['Education'].value_counts()
temp = data[['MonthlyIncome','Education']]
grps = pd.unique(temp['Education'])
d_data = {grp:data['MonthlyIncome'][data.Education == grp] for grp in grps}
#print(d_data)
stats.f_oneway(d_data[1],d_data[2],d_data[3],d_data[4],d_data[5])

```
